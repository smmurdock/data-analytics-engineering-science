# Boolean Arrays

- **boolean array** - an array of boolean values
- Sometimes also called **boolean vectors** or **boolean masks**


- `bool` values can either be `True` or `False`

An operation between an ndarray and a single value results in a new ndarray:

```python
import numpy as np
```

```python
print(np.array([2, 4, 6, 8]) + 10)
```

- The `+ 10` operation is applied to each value in the array.

### Boolean Operations

A boolean operation between an ndarray and a single value:

```python
print(np.array([2,4,6,8]) < 5)
```

### Practice

```python
# given values
a = np.array([1, 2, 3, 4, 5])
b = np.array(["blue", "blue", "red", "blue"])
c = np.array([80.0, 103.4, 96.9, 200.3])
```

1. Use vectorized boolean operations to:
    - Evaluate whether the elements in array `a` are less than `3`. Assign the result to `a_bool`.
    - Evaluate whether the elements in array `b` are equal to `"blue"`. Assign the result to `b_bool`.
    - Evaluate whether the elements in array `c` are greater than `100`. Assign the result to `c_bool`.

```python
a_bool = a < 3
b_bool = b == "blue"
c_bool = c > 100
```

```python
print(a_bool)
print(b_bool)
print(c_bool)
```
