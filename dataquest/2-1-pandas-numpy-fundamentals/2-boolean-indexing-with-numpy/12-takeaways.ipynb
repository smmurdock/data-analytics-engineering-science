### Reading CSV Files with NumPy

- Reading in a CSV file:

```python
import numpy as np
taxi = np.genfromtxt('nyc_taxis.csv', delimiter=',', skip_header=1)
```

---
### Boolean Arrays

- Creating a Boolean array from filtering criteria:

```python
np.array([2, 4, 6, 8]) < 5
```

- Boolean filtering for 1D ndarray:

```python
a = np.array([2, 4, 6, 8])
filter = a < 5
a[filter]
```

- Boolean filtering for 2D ndarray

```python
tip_amount = taxi[:, 12]
tip_bool = tip_amount > 50
top_tips = taxi[tip_bool, 5:14]
```

---
### Assigning Values

- Assigning values in a 2D ndarray using indices:

```python
taxi[28215,5] = 1
taxi[:,0] = 16
taxi[1800:1802,7] = taxi[:,7].mean()
```

- Assigning values using Boolean arrays:

```python
taxi[taxi[:,5] == 2, 15] = 1
```

## Concepts

- Selecting values from an ndarray using Boolean arrays is very powerful. Using Boolean arrays helps us think in terms of filters on the data, instead of specific index values (like in Python lists).
