```python
import numpy as np
```

```python
my_numbers = [
    [6, 5],
    [1, 3],
    [5, 6],
    [1, 4],
    [3, 7],
    [5, 8],
    [3, 5],
    [8, 4]
]
```

```python
# convert the list of lists to an ndarray
my_numbers = np.array(my_numbers)
```

```python
# select each of the columns - the result of which will be a 1D ndarray
col1 = my_numbers[:, 0]
col2 = my_numbers[:, 1]

# add the two columns
sums = col1 + col2
```

```python
print(sums)
```
