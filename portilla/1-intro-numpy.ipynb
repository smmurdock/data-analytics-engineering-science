```python
import numpy as np
```

```python
# standard python list
mylist = [1,2,3]
```

```python
type(mylist)
```

```python
# unsaved np array
# 1-dimensional vector
np.array(mylist)
```

```python
# the python list is unchanged because it
# wasn't saved into a variable
mylist
```

```python
# save the np array
myarr = np.array(mylist)
```

```python
myarr
```

```python
type(myarr)
```

```python
# python list of lists
my_matrix = [[1,2,3],[4,5,6],[7,8,9]]
```

```python
my_matrix
```

```python
# convert the list of lists into a 2-dimensional
# matrix
np.array(my_matrix)
```

```python
# numpy's `arange()` is very similar to
# python's `range()`
np.arange(0,10)
```

```python
np.arange(0,10,2)
```

```python
# if 1D, pass in an int
# default datatype is float
np.zeros(5)
```

```python
# if 2D, pass in a tuple of ints
# rows first, then columns
np.zeros((2,5))
```

```python
# specify to return ints instead of 
# default float
np.zeros((2,5),dtype=int)
```

```python
np.ones((4,4))
```

```python
# return evenly spaced numbers over a 
# specified interval
# NOT THE SAME AS np.arange()

# takes a start, stop, and number of spaces in between
np.linspace(0,10,3)
```

```python
np.linspace(0,80,11)
```

```python
# returns the same number as the num (third) 
# param
len(np.linspace(0,80,11))
```
